```python
import numpy as np
import pandas as pd

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_test.info()
```

## REPLACE ?

```python
dict_nan = {'Pekerjaan':{'?':np.nan},'Kelas Pekerja':{'?':np.nan}}
df_train = df_train.replace(dict_nan)
df_test = df_test.replace(dict_nan)
```

```python
df_train = df_train.fillna(method='bfill')
df_test = df_test.fillna(method='bfill')
```

## Replace Gaji dan Kelamin

```python
dict_gaji = {'Gaji':{'<=7jt':0,'>7jt':1}}
dict_kelamin = {'Jenis Kelamin':{'Perempuan':0,'Laki2':1}}
df_train = df_train.replace(dict_gaji)
df_train = df_train.replace(dict_kelamin)

df_test = df_test.replace(dict_kelamin)
```

## OneHotEncoder utk Status Perkawinan

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(dtype='int', sparse=False, drop='first')
ohe.fit(df_train[['Status Perkawinan']])
```

```python
kawin_train = ohe.transform(df_train[['Status Perkawinan']])
kt_col = pd.DataFrame(kawin_train, columns=['Berpisah', 'Cerai', 'Janda', 'Menikah', 'Menikah LDR'])
df_train = pd.concat([df_train,kt_col],axis = 1)
df_train = df_train.drop('Status Perkawinan', axis=1)

kawin_test = ohe.transform(df_test[['Status Perkawinan']])
kte_col = pd.DataFrame(kawin_test, columns=['Berpisah', 'Cerai', 'Janda', 'Menikah', 'Menikah LDR'])
df_test = pd.concat([df_test,kte_col],axis = 1)
df_test = df_test.drop('Status Perkawinan', axis=1)
```

## OHE Pekerjaan

```python
ohe_job = OneHotEncoder(dtype='int', sparse=False, drop='first')
ohe_job.fit(df_train[['Pekerjaan']])
```

```python
ohe_job.categories_
```

```python
job_train = ohe_job.transform(df_train[['Pekerjaan']])
jt_col = pd.DataFrame(job_train, columns=['Ekesekutif Managerial',
        'Mesin Inspeksi', 'Pembersih', 'Pemuka Agama', 'Penjaga',
        'Perbaikan Kerajinan', 'Petani', 'Sales', 'Servis Lainnya',
        'Spesialis', 'Supir', 'Tech-support', 'Tentara'])
df_train = pd.concat([df_train,jt_col],axis = 1)
df_train = df_train.drop('Pekerjaan', axis=1)

job_test = ohe_job.transform(df_test[['Pekerjaan']])
jte_col = pd.DataFrame(job_test, columns=['Ekesekutif Managerial',
        'Mesin Inspeksi', 'Pembersih', 'Pemuka Agama', 'Penjaga',
        'Perbaikan Kerajinan', 'Petani', 'Sales', 'Servis Lainnya',
        'Spesialis', 'Supir', 'Tech-support', 'Tentara'])
df_test = pd.concat([df_test,jte_col],axis = 1)
df_test = df_test.drop('Pekerjaan', axis=1)
```

## OHE Kelas Pekerja

```python
ohe_kp = OneHotEncoder(dtype='int', sparse=False, drop='first')
ohe_kp.fit(df_train[['Kelas Pekerja']])
```

```python
ohe_kp.categories_
```

```python
kp_train = ohe_kp.transform(df_train[['Kelas Pekerja']])
kpt_col = pd.DataFrame(kp_train, columns=['Pekerja Bebas Perusahaan',
        'Pemerintah Lokal', 'Pemerintah Negara', 'Pemerintah Provinsi',
        'Tanpa di Bayar', 'Tidak Pernah Bekerja', 'Wiraswasta'])
df_train = pd.concat([df_train,kpt_col],axis = 1)
df_train = df_train.drop('Kelas Pekerja', axis=1)

kp_test = ohe_kp.transform(df_test[['Kelas Pekerja']])
kpte_col = pd.DataFrame(kp_test, columns=['Pekerja Bebas Perusahaan',
        'Pemerintah Lokal', 'Pemerintah Negara', 'Pemerintah Provinsi',
        'Tanpa di Bayar', 'Tidak Pernah Bekerja', 'Wiraswasta'])
df_test = pd.concat([df_test,kpte_col],axis = 1)
df_test = df_test.drop('Kelas Pekerja', axis=1)
```
