```python
import numpy as np
import pandas as pd

df_train = pd.read_csv('train_nan.csv')
df_test = pd.read_csv('test_nan.csv')
```

```python
df_train
```

```python
from sklearn.preprocessing import StandardScaler

std_scale = StandardScaler().fit(df_train[['Jam per Minggu','Jmlh Tahun Pendidikan','Umur','Berat Akhir','Keuntungan Kapital','Kerugian Capital']])
```

```python
train_std = std_scale.transform(df_train[['Jam per Minggu','Jmlh Tahun Pendidikan','Umur','Berat Akhir','Keuntungan Kapital','Kerugian Capital']])
test_std = std_scale.transform(df_test[['Jam per Minggu','Jmlh Tahun Pendidikan','Umur','Berat Akhir','Keuntungan Kapital','Kerugian Capital']])
```

```python
df_tr_std = pd.DataFrame(train_std, columns=['Jam per Minggu','Jmlh Tahun Pendidikan','Umur','Berat Akhir','Keuntungan Kapital','Kerugian Capital'])
df_te_std = pd.DataFrame(test_std, columns=['Jam per Minggu','Jmlh Tahun Pendidikan','Umur','Berat Akhir','Keuntungan Kapital','Kerugian Capital'])
```

```python
cektr = df_train.drop(['Jam per Minggu','Jmlh Tahun Pendidikan','Umur','Berat Akhir','Keuntungan Kapital','Kerugian Capital'], axis=1)
trainkun = pd.concat([cektr,df_tr_std],axis = 1)

cekte = df_test.drop(['Jam per Minggu','Jmlh Tahun Pendidikan','Umur','Berat Akhir','Keuntungan Kapital','Kerugian Capital'], axis=1)
testkun = pd.concat([cekte,df_te_std],axis = 1)
```

```python
trainkun.to_csv('train_std.csv', index=False)
```

```python
testkun.to_csv('test_std.csv', index=False)
```
